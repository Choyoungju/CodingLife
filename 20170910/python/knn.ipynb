```python
def display_data(X, y, clf):
    x_min, x_max = X[:, 0].min(), X[:, 0].max()
    y_min, y_max = X[:, 1].min(), X[:, 1].max()

    ax = np.arange(x_min - 10, x_max + 10, 0.1)
    ay = np.arange(y_min - 10, y_max + 10, 0.1)
    vx, vy = np.meshgrid(ax, ay)

    predict_y = clf.predict(
        np.stack([vx.ravel(), vy.ravel()], axis=-1)
    )

    plt.contourf(vx, vy, predict_y.reshape(vx.shape), alpha=0.1)
    plt.scatter(X[:, 0], X[:, 1], c=y, s=1)
    plt.show()
```

```python
for i in range(1, 20):
    print('k = ', i)
    clf = KNeighborsClassifier(n_neighbors=i)
    clf.fit(X_train, y_train)
    display_data(X, y, clf)
```
