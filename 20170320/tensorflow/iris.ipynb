```python
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf

session = tf.Session()
```

```python
df = pd.read_csv('./iris.csv')
df
```

```python
x = df[['Sepal.Length', 'Sepal.Width', 'Petal.Length', 'Petal.Width']]
x -= x.min()
x /= x.max()
x
```

```python
x = x.as_matrix().astype('float32')
x
```

```python
plt.scatter(x[:, 0], x[:, 2])
plt.show()
```

```python
labels, uniques = pd.factorize(df['Species'])
labels, uniques
```

```python
plt.scatter(x[:, 0], x[:, 2], c=labels)
plt.show()
```

```python
t = session.run(tf.one_hot(labels, depth=len(uniques)))
t
```

```python
plt.scatter(x[:, 0], x[:, 2], c=session.run(tf.argmax(t, axis=1)))
plt.show()
```

```python
features = [0, 1, 2, 3]
k = len(uniques)

w = tf.Variable(tf.truncated_normal([len(features), k]))
b = tf.Variable(tf.truncated_normal([k]))
y = tf.nn.softmax(tf.matmul(x[:, features], w) + b)

cross_entropy = tf.reduce_sum(-t * tf.log(y), axis=1)
loss = tf.reduce_mean(cross_entropy)
train = tf.train.GradientDescentOptimizer(0.5).minimize(loss)
```

```python
correct = tf.equal(tf.argmax(t, 1), tf.argmax(y, 1))
accuracy = tf.reduce_mean(tf.cast(correct, tf.float32))
```

```python
session.run(tf.global_variables_initializer())

def draw(index):
    print(index, session.run(accuracy))
    c = session.run(tf.argmax(y, axis=1))
    plt.scatter(x[:, 0], x[:, 2], c=c)
    plt.show()

draw(0)
for i in range(20):
    for j in range(10):
        session.run(train)
    draw(i + 1)
```
