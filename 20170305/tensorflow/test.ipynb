```python
# Import

import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelBinarizer
from sklearn.model_selection import train_test_split
```

```python
# Read CSV

df = pd.read_csv('iris.csv')
df
```

```python
# Data Types

df.dtypes
```

```python
# Rename Columns

df.columns = [x.lower().replace('.', '_') for x in df.columns]
df.head()
```

```python
# Category Factorize

labels, uniques = pd.factorize(df.species)
df['label'] = labels
df[['species', 'label']]
```

```python
# Draw Pie Chart

df.species.value_counts().plot.pie(autopct='%d')
plt.show()
```

```python
# Draw Chart

df.plot.scatter(x='sepal_length', y='sepal_width', c='label')
plt.show()

colors = 'rgb'
ax = None
for i in range(3):
    ax = df[df.label == i].plot.scatter('sepal_length', 'sepal_width', c=colors[i], ax=ax)
plt.show()

df.plot.scatter('sepal_length', 'sepal_width', c=pd.Series(['r', 'g', 'b'])[df.label])
plt.show()
```

```python
# Column for Plot

df['label_color'] = pd.Series(['r', 'g', 'b'])[df.label].values
df[['label', 'label_color']]
```

```python
# Draw Charts

from itertools import combinations

for x, y in combinations(['sepal_length', 'sepal_width', 'petal_length', 'petal_width'], 2):
    df.plot.scatter(x, y, c=df.label_color)
    plt.show()
```

```python
# Split Data Set into Train & Test

train, test = train_test_split(df, test_size=0.2)

train
```

```python
# Features

def column(df, name):
    m = df.as_matrix([name])
    return (m - m.min()) / m.ptp()

train_features = np.hstack([column(train, 'sepal_length'), column(train, 'petal_length')])
test_features = np.hstack([column(test, 'sepal_length'), column(test, 'petal_length')])

train_features
```

```python
# Label-Class

lb = LabelBinarizer()
lb.fit(df.label)

train_labels = lb.transform(train.label).astype('float')
test_labels = lb.transform(test.label).astype('float')

train_labels
```

```python
# Softmax Regression

k = lb.classes_.size

x = tf.placeholder(tf.float32, [None, 2])
y = tf.placeholder(tf.float32, [None, 3])

w = tf.Variable(tf.truncated_normal([2, k], stddev=0.1))
b = tf.Variable(tf.truncated_normal([k], stddev=0.1))
y_ = tf.nn.softmax(tf.matmul(x, w) + b)

# Loss Function

cross_entropy = -tf.reduce_sum(y * tf.log(y_), axis=1)
loss = tf.reduce_mean(cross_entropy)

# Gradient Descent Optimization

learning_rate = 0.5
optimizer = tf.train.GradientDescentOptimizer(learning_rate)
```

```python
# Initialize

session = tf.InteractiveSession()
tf.global_variables_initializer().run()

# Train & Test

train_data = {x: train_features, y: train_labels}
test_data = {x: test_features, y: test_labels}

correct_prediction = tf.equal(tf.argmax(y, 1), tf.argmax(y_, 1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))

def draw(index):
    print(index)
    print(session.run(accuracy, feed_dict=train_data))
    print(session.run(accuracy, feed_dict=test_data))
    labels = session.run(tf.argmax(y_, 1), feed_dict=train_data)
    train.plot.scatter('sepal_length', 'petal_length', c=pd.Series(['r', 'g', 'b'])[labels])
    plt.show()

draw(0)

for i in range(20):
    optimizer.minimize(loss).run(feed_dict=train_data)
    draw(i + 1)
```
